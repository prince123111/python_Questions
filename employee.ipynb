# How Many Employee in Various Department ?

```python
y=x["department"].value_counts().reset_index()
y=y.sort_values(by="department",ascending=False)
y
```

```python
plt.figure(figsize=(20,8))
sns.barplot(data=y.head(20),x='index',y='department')
plt.xticks(rotation=60)
plt.title("Counting the Department",fontweight="bold",fontsize=20,color="black")
plt.show()
```

* We can clearly see that maximum number of employees are Analyst. Number of employee are in analyst department 390.
* And After Finance Department in second  most employee whose Number is 355. so which jobs on trending easily we understand
with the help of the barplot 

# Count How many employee in Various City ?

```python
z=x['city'].value_counts().reset_index().sort_values(by="city",ascending=False)
z
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=z.head(15),x="index",y="city",palette="Accent")
plt.xticks(rotation=60)
plt.show()
```
