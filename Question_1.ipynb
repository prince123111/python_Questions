### Question :-
You have given dataset in sentences if sentences in  ['excited','happy','party','good'] then fill Casual and 
that sentences have ['rejects','bullied','lonely','dead','wrong'] then fill Dipressed else fill None ?

## Let's Do that :-

```python
import pandas as pd
import numpy as np
```

```python
x=pd.read_csv("C:/Users/Prince kumar/Downloads/Copy of Sentences.xlsx - Sheet1.csv")
x.drop("opt",axis=1,inplace=True)
x
```

```python
ca=['excited','happy','party','good']
de=['rejects','bullied','lonely','dead','wrong']
y=[]
z=[]
```

```python
for i in x['Sentences']:
	for j in ca:
		if j in i:
			y.append("Casual")
			break
	else:
		for k in de:
			if k in i:
				y.append("Depressed")
				break
		else:
			y.append("None")

x['Mood']=y
```
